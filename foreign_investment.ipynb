```python
# Data Cleaning, Data Pre-processing, Data Analysis

import pandas as pd
import numpy as np
investment = pd.read_csv("/content/FDI data.csv")      #copy the file path
investment.head()                                               #Reading Dataset
```

```python
# Total Rows and Columns

investment.shape
```

**Data Cleaning**

```python
# All Columns

columns = list(investment)
```

```python
# Checking Null Values

investment.isnull().sum()
```

```python
# Calculating Total Zeros

(investment[columns[0:19]] == 0).sum()
```

```python
# Replace '0' with 'Null'
investment[columns[1:19]] = investment[columns[1:19]].replace(0, np.nan)
```

```python
# Again Checking Null Values

investment.isnull().sum()
```

Data Cleaning is done. All zero values are replaced with Null.
